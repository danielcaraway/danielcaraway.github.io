## STEP 4: Clean the data

```python
output_filename = 'output.txt'
outfile = open(output_filename, 'w')
outfile.close()
outfile = open(output_filename, 'a')

def clean_word(word):
    word=word.lower()
    word=word.lstrip()
    word=word.lstrip("\\n")
    word=word.strip("\n")
#     word=word.strip("\\n")
    word=word.replace(",","")
    word=word.replace(" ","")
    word=word.replace("_","")
    word=re.sub('\+', ' ',word)
    word=re.sub('.*\+\n', '',word)
    word=re.sub('zz+', ' ',word)
    word=word.replace("\t","")
    word=word.replace(".","")
    #word=word.replace("\'s","")
    word=word.strip()
    word = word.replace("\\'","")
    if word not in ["", "\\", '"', "'", "*", ":", ";"]:
        if len(word) >= 3:
            if not re.search(r'\d', word): ##remove digits
                return word

clean_filename = "clean_movies_v2.txt"
cleanfile = open(clean_filename, 'w')
cleanfile.close()
cleanfile = open(clean_filename, 'a')

def display_rows(file_data):

    for row in file_data:
        
        row = row.lstrip()
        row = row.rstrip()
        row = row.strip()
        raw_row = "\n\nROW:" + row + "\n"
        outfile.write(raw_row)
        row_list = row.split(" ")
        new_list = []
        for word in row_list:
            to_put_in_outfile = "The next word BEFORE is: "+ word +"\n"
            outfile.write(to_put_in_outfile)
            word = clean_word(word)
#             print(word)
            if word:
                new_list.append(word)
#         label = new_list[-1]
        label = ''.join(char for char in new_list[-1] if char.isalpha())
        new_list.pop()
        just_text = ' '.join(new_list)
#         print(label)
#         print(just_text)
        to_write = label + ',' + just_text + '\n'
        print('TO WRITE', to_write)
        cleanfile.write(to_write)
#     clean_file.close()
        
#         print('*************')
        
#             print(word)
            
new_list = display_rows(file_data)
file.close()
cleanfile.close()
```

```python
data = pd.read_csv('clean_movies_v2.txt', engine='python-fwf', sep=",")
data
```

```python
data['label'] = data.apply(lambda x: x['textreviewclass'].split(',')[0], axis=1)
data['review'] = data.apply(lambda x: x['textreviewclass'].split(',')[1], axis=1)
data = data.drop('textreviewclass', axis=1)
data
```

```python
# GATES WAY
data_nolabel = data.drop('label', axis=1)
data_label = data['label']
all_reviews=[]
for i in range(0,len(data_nolabel)):
    review = data_nolabel.iloc[i,0]
    all_reviews.append(review)
# print(all_reviews[1:4])

# KENDRA'S WAY 
test = data['review'].tolist()
test[1:4]
```

```python
## "content" means that you will need a LIST that
## contains all the text. In other words, the first element in

# This line will change based on our project
input_data = all_reviews
vectorizer = CountVectorizer(input="content")

vectorized_count = vectorizer.fit_transform(input_data)
print(vectorized_count)
```

```python
column_names = vectorizer.get_feature_names()
vectorized_text = pd.DataFrame(vectorized_count.toarray(),columns = column_names)
print(vectorized_text)
```
